```python
'''problem statement 2: THree students choose elective subjects.
find the common subjects chlosen by all students using set operation
concept used : set intersection (&)'''

student1 = {"Maths", "Physics", "Computer Science"}
student2 = {"Computer Science", "Maths", "Biology"}
student3 = {"Maths", "Computer Science", "Chemistry"}
common_subjects = student1 & student2 & student3
print("Common elective subjects chosen by all students:")
print(common_subjects)
```

```python
'''problem statement 3: one set contains student in the cricket club and another in football club.
find student who are :1 in both clubs 2 in only one club
concept used : set intersection (|) and set intersection (&)'''

cricket = {"Amit", "Binod", "Charle", "Darsh"}
football = {"Charle", "Darsh", "Elephant", "Finn"}
both_clubs = cricket & football
only_one_club = (cricket | football) - both_clubs
print("Students in both clubs:", both_clubs)
print("Students in only one club:", only_one_club)
```
